```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')

vec_king = wv['king']
```

Gensim is a popular Python library for Natural Language Processing (NLP).

It is widely used for word embeddings (vector representations of words).

Word2Vec is a model that learns word embeddings from text.

KeyedVectors is used for loading pre-trained embeddings without retraining.

gensim.downloader lets us easily load pre-trained models.



This loads the famous Google News Word2Vec model trained on 100 billion words.

"300" means each word is represented as a 300-dimensional vector.

Instead of training on your own text, you use this huge pre-trained model.

```python

vec_king
```

```python
vec_king.shape
```

wv['king'] gives the vector representation of the word "king".

This is a NumPy array of size (300,) → a 300-dimensional word embedding.

These embeddings capture semantic meaning (similar words are close in vector space).

```python
wv['cricket']
```

This gives the vector for "cricket".

Internally, the numbers represent the position of the word in semantic space.

```python
wv.most_similar('cricket')
```

Returns a list of words most similar to "cricket" (like "baseball", "soccer", "hockey").

This works because Word2Vec places related words close together in the vector space.

```python

wv.most_similar('happy')
```

Gives synonyms/related words like "joyful", "cheerful", etc.

```python
wv.similarity("hockey","sports")
```

Computes cosine similarity between the vectors of "hockey" and "sports".

Value ranges from -1 to 1:

Close to 1 → very similar

Close to 0 → unrelated

Negative → opposite meanings (rare in embeddings).

```python
vec=wv['king']-wv['man']+wv['woman']
```

```python

vec
```

This is the analogy trick:

"king" is related to "man" in the same way that "queen" is related to "woman".

So king - man + woman ≈ queen.


→ You’ll get “queen” at the top.
